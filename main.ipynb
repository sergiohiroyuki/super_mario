```python
import gym_super_mario_bros
from nes_py.wrappers import JoypadSpace
from gym_super_mario_bros.actions import RIGHT_ONLY
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import CheckpointCallback
from gym import Wrapper, ObservationWrapper
import gym
import numpy as np
import cv2
import time

# Wrapper personalizado que ignora argumentos inesperados
class CustomSeedWrapper(Wrapper):
    def reset(self, **kwargs):
        kwargs.pop('seed', None)
        kwargs.pop('options', None)
        return self.env.reset(**kwargs)

# WarpFrame Wrapper para normalização de frames (preprocessamento)
class WarpFrame(ObservationWrapper):
    def __init__(self, env, width=84, height=84, grayscale=True):
        super().__init__(env)
        self.width = width
        self.height = height
        self.grayscale = grayscale
        num_colors = 1 if grayscale else 3
        new_space = gym.spaces.Box(
            low=0,
            high=255,
            shape=(self.height, self.width, num_colors),
            dtype=np.uint8,
        )
        self.observation_space = new_space

    def observation(self, obs):
        if self.grayscale:
            obs = cv2.cvtColor(obs, cv2.COLOR_RGB2GRAY)
        obs = cv2.resize(obs, (self.width, self.height), interpolation=cv2.INTER_AREA)
        if self.grayscale:
            obs = np.expand_dims(obs, -1)
        return obs

# Atualizando a função cria_env para usar os wrappers personalizados
def cria_env(nome_env):
    def _init():
        env = gym_super_mario_bros.make(nome_env)
        env = JoypadSpace(env, RIGHT_ONLY)
        env = CustomSeedWrapper(env)  # Wrapper para remover argumentos inesperados
        env = WarpFrame(env)  # Wrapper para normalização de frames
        return env
    return _init

# Utilizando DummyVecEnv para múltiplas instâncias
threads_cpu = 8  # Número de processos

# Usando DummyVecEnv com o ambiente criado
env = DummyVecEnv([cria_env(nome_env='SuperMarioBros-1-1-v0') for _ in range(threads_cpu)])

# Configurando o modelo PPO
modelo = PPO(policy='CnnPolicy', env=env, gamma=0.99, n_steps=1024, batch_size=64, learning_rate=0.0003, vf_coef=0.5, ent_coef=0.01, verbose=1)

# Configurando o callback para checkpoints
checkpoint_callback = CheckpointCallback(save_freq=6250, save_path='C:/Users/hiroy/super_mario_2', name_prefix='n1024b64l3')

# Treinamento do modelo
inicio = time.time()
modelo.learn(total_timesteps=10000000, log_interval=10, callback=checkpoint_callback)
fim = time.time()

print('Tempo total do treinamento em horas:', (fim - inicio) / 3600)
```

```python
import gym
from gym import Wrapper
from gym_super_mario_bros import make
from nes_py.wrappers import JoypadSpace
from gym_super_mario_bros.actions import RIGHT_ONLY
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import PPO
import time

# Definindo um wrapper para sobrescrever o método reset()
class RemoveSeedResetWrapper(Wrapper):
    def reset(self, **kwargs):
        if 'seed' in kwargs:
            kwargs.pop('seed')
        return self.env.reset(**kwargs)

# Função para criar o ambiente
def cria_env(nome_env):
    def _init():
        env = make(nome_env)  # Criando o ambiente Mario
        env = JoypadSpace(env, RIGHT_ONLY)  # Limitando ações para direita
        env = RemoveSeedResetWrapper(env)  # Aplicando o wrapper que remove o 'seed'
        # env = nosso_wrapper(env)  # Remova ou ajuste se estiver usando um wrapper personalizado
        return env
    return _init

# Criando o ambiente com DummyVecEnv
env = DummyVecEnv([cria_env(nome_env='SuperMarioBros-1-1-v0')])

# Configurando o modelo PPO
modelo = PPO(
    policy='CnnPolicy',
    env=env,
    gamma=0.99,
    n_steps=1024,
    batch_size=64,
    learning_rate=0.0003,
    vf_coef=0.5,
    ent_coef=0.01,
    verbose=1
)

# Treinamento do modelo
inicio = time.time()
modelo.learn(total_timesteps=100000)  # Número de timesteps reduzido para testes
fim = time.time()

print('Tempo total do treinamento em horas:', (fim - inicio) / 3600)
```

```python
import numpy as np
import os
os.environ.setdefault('PATH', '')
from collections import deque
import sys
sys.path.append('C:/Users/hiroy/super_mario_2')
import gym
from gym import spaces
import cv2
cv2.ocl.setUseOpenCL(False)
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import CheckpointCallback
from stable_baselines3.common.vec_env import DummyVecEnv
from gym.wrappers import FrameStack
import time

# Definição do MaxAndSkipEnv
class MaxAndSkipEnv(gym.Wrapper):
    def __init__(self, env, skip=4):
        super().__init__(env)
        self._skip = skip
        self._obs_buffer = deque(maxlen=2)

    def step(self, action):
        total_reward = 0.0
        done = False
        for _ in range(self._skip):
            obs, reward, done, info = self.env.step(action)  # Remove o truncated
            self._obs_buffer.append(obs)
            total_reward += reward
            if done:
                break
        max_frame = np.max(np.stack(self._obs_buffer), axis=0)
        return max_frame, total_reward, done, info  # Remove o truncated

    def reset(self, **kwargs):
        self._obs_buffer.clear()
        obs = self.env.reset(**kwargs)  # Remove o info aqui
        self._obs_buffer.append(obs)
        return obs  # Remove o info

# Definição do WarpFrame
class WarpFrame(gym.ObservationWrapper):
    def __init__(self, env, width=84, height=84, grayscale=True):
        super().__init__(env)
        self.width = width
        self.height = height
        self.grayscale = grayscale
        if self.grayscale:
            self.observation_space = gym.spaces.Box(low=0, high=255, shape=(self.height, self.width, 1), dtype=np.uint8)
        else:
            self.observation_space = gym.spaces.Box(low=0, high=255, shape=(self.height, self.width, 3), dtype=np.uint8)

    def observation(self, frame):
        if self.grayscale:
            frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        frame = cv2.resize(frame, (self.width, self.height), interpolation=cv2.INTER_AREA)
        if self.grayscale:
            frame = np.expand_dims(frame, axis=-1)
        return frame

# Definição do ClipRewardEnv
class ClipRewardEnv(gym.RewardWrapper):
    def reward(self, reward):
        """ Clip the reward to be between -1 and 1. """
        return np.clip(reward, -1, 1)

# Outros wrappers (NoopResetEnv etc.)

def nosso_wrapper(env):
    env = MaxAndSkipEnv(env, skip=4)
    env = WarpFrame(env)
    env = FrameStack(env, 4)
    env = ClipRewardEnv(env)  # Agora ClipRewardEnv está definido
    return env

# Ambiente do Mario
env_id = "SuperMarioBros-v0"  # Certifique-se que essa é a versão correta do ambiente

# Alteração para DummyVecEnv (um único processo)
env = DummyVecEnv([lambda: nosso_wrapper(gym.make(env_id))])

# Checkpoint para salvar o progresso
checkpoint_callback = CheckpointCallback(save_freq=10000, save_path='./checkpoints/', name_prefix='rl_model')

# Criação do modelo PPO
modelo = PPO(policy='MlpPolicy', env=env, gamma=0.99, n_steps=1024, batch_size=64, learning_rate=0.0003, vf_coef=0.5, ent_coef=0.01, verbose=1)

# Iniciando o treinamento com medição do tempo
inicio = time.time()
modelo.learn(total_timesteps=10000000, log_interval=10, callback=checkpoint_callback)
fim = time.time()

# Tempo total do treinamento
print('Tempo total do treinamento em horas:', (fim-inicio)/3600)
```
